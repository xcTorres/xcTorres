# Data Operation 

```python
import torch

torch.manual_seed(0)
torch.cuda.manual_seed(0)
print(torch.__version__)
```

## Create Tensor

```python
x = torch.empty(5, 3)
print(x)
```

```python
x = torch.rand(5, 3)
print(x)
```

```python
x = x.new_ones(5, 3, dtype=torch.float64)      # 返回的tensor默认具有相同的torch.dtype和torch.device
print(x)

x = torch.randn_like(x, dtype=torch.float)    # 指定新的数据类型
print(x)
```

```python
print(x.size())
print(x.shape)
```

```python
y = torch.rand(5, 3)
print(x + y)
```
